```python
import pandas as pd
segments_df = pd.read_csv("agora/segments.csv")
documents_df = pd.read_csv("agora/documents.csv")

documents_df = documents_df[
    [
        "AGORA ID",
        "Official name",
        "Authority",
        "Collections",
        "Short summary"
    ]
]
merged_df = segments_df.merge(
    documents_df,
    left_on="Document ID",
    right_on="AGORA ID",
    how="left"
)



merged_df.head(5)

```
